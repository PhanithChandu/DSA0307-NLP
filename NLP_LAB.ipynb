#EXP 1 - Write program demonstrates how to use regular expressions in Python to match and search for patterns in text.

```python
import re

text = "My email is student123@gmail.com and my phone number is 9876543210."

email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'
email = re.search(email_pattern, text)

if email:
    print("Email found:", email.group())
else:
    print("Email not found.")

match = re.match(r"My", text)

if match:
    print("Matched at the beginning:", match.group())
else:
    print("No match at the beginning.")

phone_pattern = r'\b\d{10}\b'
phone = re.search(phone_pattern, text)

if phone:
    print("Phone number found:", phone.group())
else:
    print("Phone number not found.")
```

# EXP 2 - Implement a basic finite state automaton that recognizes a specific language or pattern. In this example, we'll create a simple automaton to match strings ending with 'ab' using python.



```python
def fsa(string):
    state = 0

    for ch in string:
        if state == 0:
            if ch == 'a':
                state = 1
            else:
                state = 0
        elif state == 1:
            if ch == 'a':
                state = 1
            elif ch == 'b':
                state = 2
            else:
                state = 0
        elif state == 2:
            if ch == 'a':
                state = 1
            else:
                state = 0

    if state == 2:
        print("Accepted")
    else:
        print("Rejected")

s = input("Enter a string: ")
fsa(s)
```

#EXPP 03 - Write program demonstrates how to perform morphological analysis using the NLTK library in Python.


```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

nltk.download('punkt_tab')

text = "The students are studying machine learning and developing intelligent systems."

tokens = word_tokenize(text)

stemmer = PorterStemmer()

print("Original Words:")
print(tokens)

print("\nStemmed Words:")
for word in tokens:
    print(word, "->", stemmer.stem(word))
```

EXP 04 - Implement a finite-state machine for morphological parsing. In this example, we'll create a
simple machine to generate plural forms of English nouns using python.


```python
def pluralize(noun):
    if noun.endswith(("s", "x", "z", "ch", "sh")):
        return noun + "es"
    elif noun.endswith("y") and noun[-2] not in "aeiou":
        return noun[:-1] + "ies"
    else:
        return noun + "s"

word = input("Enter a noun: ")
print("Plural form:", pluralize(word))
```

#EXP 06 - Implement a basic N-gram model for text generation. For example, generate text using a bigram model using python.

```python
import random

text = "the cat sat on the mat the cat ate the fish"

words = text.split()

bigrams = {}

for i in range(len(words) - 1):
    key = words[i]
    value = words[i + 1]
    if key not in bigrams:
        bigrams[key] = []
    bigrams[key].append(value)

current = random.choice(words)
generated = [current]

for _ in range(9):
    if current in bigrams:
        current = random.choice(bigrams[current])
        generated.append(current)
    else:
        break

print("Generated Text:")
print(" ".join(generated))
```

#EXP 08 -Implement a simple stochastic part-of-speech tagging algorithm using a basic probabilistic model to assign POS tags using python.

```python
import random

pos_model = {
    "I": ["PRP"],
    "like": ["VBP", "IN"],
    "to": ["TO"],
    "eat": ["VB", "NN"],
    "apple": ["NN"],
    "today": ["NN", "RB"]
}

sentence = input("Enter a sentence: ").split()

print("POS Tags:")
for word in sentence:
    if word in pos_model:
        tag = random.choice(pos_model[word])
    else:
        tag = "NN"
    print(word, ":", tag)
```

#EXP 09 - Implement a rule-based part-of-speech tagging system using regular expressions using python.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import RegexpTagger

nltk.download('punkt')

patterns = [
    (r'.*ing$', 'VBG'),
    (r'.*ed$', 'VBD'),
    (r'.*es$', 'VBZ'),
    (r'.*ly$', 'RB'),
    (r'.*ous$', 'JJ'),
    (r'.*able$', 'JJ'),
    (r'.*ness$', 'NN'),
    (r'.*ment$', 'NN'),
    (r'.*s$', 'NNS'),
    (r'.*', 'NN')
]

tagger = RegexpTagger(patterns)

text = "The students are studying carefully and completed the assignment"

words = word_tokenize(text)
tagged = tagger.tag(words)

print("Rule-Based POS Tags:")
for word, tag in tagged:
    print(word, ":", tag)
```

#EXP 10 - Implement transformation-based tagging using a set of transformation rules, apply a simple rule to tag words using python.

```python
sentence = "The cat is running quickly"

words = sentence.split()

tags = []

for word in words:
    if word.lower() in ["the", "a", "an"]:
        tags.append((word, "DT"))
    elif word.lower() in ["is", "am", "are", "was", "were"]:
        tags.append((word, "VB"))
    else:
        tags.append((word, "NN"))

transformed_tags = []

for word, tag in tags:
    if word.endswith("ing"):
        tag = "VBG"
    elif word.endswith("ly"):
        tag = "RB"
    transformed_tags.append((word, tag))

print("Tagged Sentence:")
for word, tag in transformed_tags:
    print(word, ":", tag)
```
